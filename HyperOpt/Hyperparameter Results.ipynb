```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
def show_hp(df, eval_type='val_acc', PLOT=True):
    row = df.iloc[df[eval_type].idxmax()]
    ID  = row['Trial-ID']
    
    for hp in ['Trial-ID', 'model', 'num_layers', 'lr', 'Iteration', 'val_acc']:
        print hp,':',row[hp]
        
    if PLOT:
        y = df[df['Trial-ID'] == ID][eval_type]
        x = df[df['Trial-ID'] == ID]['Iteration']
        plt.plot(x,y)
        plt.show()
```

# MNIST

```python
mnist = pd.read_csv('SherpaResults/mnist/results.csv')
mnist = mnist[mnist['Status'] != 'COMPLETED']
```

```python
mnist_fcn = mnist[mnist['model'] == 'fcn'].reset_index()
mnist_cnn = mnist[mnist['model'] == 'cnn'].reset_index()

# sns.lineplot(x='Iteration', y='val_acc', data=mnist_fcn, label='fcn')
# sns.lineplot(x='Iteration', y='val_acc', data=mnist_cnn, label='cnn')
# plt.legend(); plt.show()
```

```python
show_hp(mnist_cnn)
```

```python
show_hp(mnist_fcn)
```
